# べき乗とネイピア数

## ● べき乗
同じ数を何度か掛け合わせることを、**べき乗**といいます。  
例えば、
$$3\times 3\times 3\times 3\times 3$$
は、3を5つかけていますが、次のように短く表すことができます。

$$3^5$$

以下のコードは、次のべき乗の式をグラフに描画します。
$$ y = 2^x $$
$$y = 3^x$$

#### 数式の練習用
$$$$
$$$$

```
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2, 2)

y_2 = 2**x
y_3 = 3**x

plt.plot(x, y_2, label="2^x")
plt.plot(x, y_3, label="3^x")
plt.legend()

plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.grid()
plt.show()
```

## ● ネイピア数

以下のコードは、ネイピア数のべき乗の式をグラフで描画します。

$$y=e^x$$

#### 数式の練習用
$$y=e^x$$

```
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2, 2)
e = np.e  # ネイピア数

y_2 = 2**x  # 比較用
y_e = e**x  # ネイピア数のべき乗
y_3 = 3**x  # 比較用

plt.plot(x, y_2, label="2^x")
plt.plot(x, y_e, label="e^x")
plt.plot(x, y_3, label="3^x")
plt.legend()

plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.grid()
plt.show()
```

```
# コード練習用
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2, 2)
e = np.e

y_2 = 2**x
y_e = e**x
y_3 = 3**x

plt.plot(x, y_2, label="2^x")
plt.plot(x, y_e, label="e^x")
plt.plot(x, y_3, label="3^x")


plt.xlabel("x", size=14)
plt.ylabel("y", size=14)

plt.legend()
plt.grid()
plt.show()
```

 ネイピア数のべき乗には、  
   
**yの微小な変化 / xの微小な変化**  

すなわち曲線の傾きが元の関数に近づいていくという性質があります。  
以下のコードでは、これをグラフで描画することにより確認します。  

```
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-2, 2)
dx = 0.1  # xの微小な変化
e = np.e  # ネイピア数

y_e = e**x  # 元のべき乗の関数
y_de = (e**(x+dx) - e**x) / dx  # yの微小な変化/xの微小な変化

plt.plot(x, y_e, label="e^x")
plt.plot(x, y_de, label="de")
plt.legend()

plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.grid()
plt.show()
```

```
# コード練習用
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-2, 2)
dx = 0.01
e = np.e

y_e = e**x
y_de = (e**(x+dx) - e**x) / dx

plt.plot(x, y_e, label="e^x")
plt.plot(x, y_de, label="de")
plt.legend()

plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.grid()
plt.show()
```
